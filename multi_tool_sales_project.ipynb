```python
#### removal of outliers 
def removing_outliers(data,column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3-Q1 
    return(data[(data[column]>=Q1-1.5*IQR) & (data[column]<=Q3+1.5*IQR)])
order_centric = removing_outliers(order_centric,'TotalAmount')
order_centric = removing_outliers(order_centric,'Revenue')
order_centric = removing_outliers(order_centric,'Quantity')
order_centric = removing_outliers(order_centric,'Cost')
order_centric = removing_outliers(order_centric,'Cst per order')
order_centric = removing_outliers(order_centric,'Profit')
order_centric
```

```python
#### 19) Divide the customers into groups based on Recency, Frequency, and Monetary (RFM Segmentation) -
####Divide the customers into Premium, Gold, Silver, Standard customers and understand the behaviour of each segment of customers

max_date = order_centric['OrderDate'].max()
RFM = order_centric.groupby('CustomerId').agg(last_date = ('OrderDate','max'),Frequency = ('OrderId','count'),Monetary = ('Revenue','sum')).reset_index()
RFM['Recency'] = (max_date - RFM['last_date']).dt.days
RFM = RFM.sort_values(by= ['Recency','Frequency','Monetary'],ascending = [True,False,False])
RFM['R_Score'] = pd.qcut(RFM['Recency'],4,labels=[1,2,3,4])
RFM['F_Score'] = pd.qcut(RFM['Frequency'].rank(method ='first'),4,labels=[4,3,2,1])
RFM['M_Score'] = pd.qcut(RFM['Monetary'].rank(method ='first'),4,labels=[4,3,2,1])
RFM['RFM_SCORE'] = RFM[['R_Score','F_Score','M_Score']].astype('int').sum(axis = 1)
RFM['Segments'] = pd.cut(RFM['RFM_SCORE'],bins= [0,3,6,9,12],labels = ['Standard','Silver','Gold','Premium'])
```

```python
Segments_counts = RFM['Segments'].value_counts().sort_index()
plt.figure(figsize =(10,6))
for Segments,group in RFM.groupby('Segments'):
    plt.hist(group['RFM_SCORE'],label = Segments,bins = 5)
plt.legend()
plt.title('Categorizing Customer based on RFM_SCORE')
```

## HYPOTHESIS TESTING 

#### DO UNIT PRICE HAVING A NEGATIVE IMPACT ON REVENUE 

```python
 
### Null Hypothesis(H0):
# No significant relationship between Revenue and UnitPrice_orders_items

#### Alternative Hypothesis (Ha):
# There is a negative relationship between Revenue and UnitPrice_orders_items

### CI & Pvalue 
## CI = 95% 
## Pvalue = 0.05

### Test - personr

### Business Conclusion 
## Here significantly the (P_value <0.05) but the correlation is positive +0.45713287583009266 here we will reject our null hypothesis that there is 
## no relationship but we will also reject the assumption that there is a negative relationship that if my 'UnitPrice_orders_items' increases then my 
## revenue decreases -- maybe beacause the quantity is not dropping sharply as unit_price increase

### Rejection Rules 
## p_value is less than 0.05 so we will reject the null hypothesis 
```

```python
import scipy.stats as stats
corr,p_value = stats.pearsonr(order_centric['UnitPrice_orders_items'],order_centric['Revenue'])
print('Correlatioin :',corr)
print('p_value :',p_value)                           
```

### Do New_Customer spend less than old customer 

```python
### Null Hypothesis(H0):
# No significant difference in revenue by differenct customer_type

#### Alternative Hypothesis (Ha):
# There is a significant difference in revenue by differenct customer_type

### CI & Pvalue 
## CI = 95% 
## Pvalue = 0.05

### Test - f_oneway()

### Business Conclusion 
## Here significantly the (P_value <0.05) here we will reject our null hypothesis that there is  no difference in reveneu by differenct customer_type 
## and there is significant difference in revenue by customer type and by mean we can see that average mean of old customer is higher than new 

### Rejection Rules 
## p_value is less than 0.05 so we will reject the null hypothesis 
```

```python
new_cust = order_centric[order_centric['Customer_Type']=='New']['Revenue']
old_cust = order_centric[order_centric['Customer_Type']=='Old']['Revenue']
stats.f_oneway(new_cust,old_cust)
```

```python
print('Average_mean_new: ',new_cust.mean())
print('Average_mean_old: ',old_cust.mean())
```

## Revenue and profit on weekend will be higher than weekdays

```python
### Null Hypothesis(H0):
# No significant difference in revenue by Days

#### Alternative Hypothesis (Ha):
# There is a significant difference in revenue by Days 

### CI & Pvalue 
## CI = 95% 
## Pvalue = 0.05

### Test - f_oneway()

### Business Conclusion 
## Here significantly the (P_value > 0.05) so we can say that days are not making as impact in the revenue 
## yead there is a little bit of a difference in revenue by days but not that significant that is so low 

### Rejection Rules 
## here p_value > 0.05 so we fail to reject the null hypothesis that the days has a significat impact in revenue
```

```python
weekdays_revenue = order_centric[order_centric['Days']=='Weekdays']['Revenue']
weekend_revenue= order_centric[order_centric['Days']=='Weekend']['Revenue']

f_stats,p_value = stats.f_oneway(weekdays_revenue,weekend_revenue)
print('f_stats: ' ,f_stats)
print('p_value: ',p_value)
```
